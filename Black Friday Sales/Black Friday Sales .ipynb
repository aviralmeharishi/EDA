## **Null Value Treatment**

```python
data.isnull().sum()
```

- So we Have Null Values in Product_Category_2 and Product_Category_3 
- We Will Use SimpleImputer for that matter as both are categorical nature so we use mode

```python
from sklearn.impute import SimpleImputer
```

```python
cols_need_to_impute = ['Product_Category_2', 'Product_Category_3']
```

```python
Si = SimpleImputer(strategy='most_frequent')

for i in cols_need_to_impute:
    data[i] = Si.fit_transform(data[[i]])
```

```python
data.isnull().sum().sum()
```

Inferences:
- Hence all the Null values are treated


## **Outlier Treatment**
